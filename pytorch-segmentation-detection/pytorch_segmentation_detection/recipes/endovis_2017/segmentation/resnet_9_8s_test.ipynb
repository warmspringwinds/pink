```python
%matplotlib inline

import sys, os
sys.path.append("/home/daniil/repos/pytorch-segmentation-detection/")
sys.path.insert(0, '/home/daniil/repos/pytorch-segmentation-detection/vision/')

# Use second GPU -pytorch-segmentation-detection- change if you want to use a first one
os.environ["CUDA_VISIBLE_DEVICES"] = '1'


from PIL import Image
from matplotlib import pyplot as plt

import torch
from torchvision import transforms
from torch.autograd import Variable
import pytorch_segmentation_detection.models.resnet_dilated as resnet_dilated

import numpy as np


img_path = '/home/daniil/datasets/data/instrument_dataset_1/left_frames/frame002.png'

valid_transform = transforms.Compose(
                [
                    transforms.CenterCrop((1024, 1280)),
                    # Just to measure time correctly -- this transform actually
                    # distorts image in a non-isotropic way
                    transforms.Resize(size=(512, 512)),
                     transforms.ToTensor(),
                     transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
                ])

img_not_preprocessed = Image.open(img_path).convert('RGB')#.resize((512, 512))

img = valid_transform(img_not_preprocessed)

img = img.unsqueeze(0)

img = Variable(img.cuda())

fcn = resnet_dilated.Resnet9_8s(num_classes=2)
fcn.load_state_dict(torch.load('/home/daniil/projects/adaptive_segmentation/resnet_9_8s.pth'))
fcn.cuda()
fcn.eval()

res = fcn(img)

_, tmp = res.squeeze(0).max(0)

segmentation = tmp.data.cpu().numpy().squeeze()

plt.imshow(img_not_preprocessed)
plt.show()

plt.imshow(segmentation)
plt.show()
```

```python
def benchmark_fcn():
    
    img = valid_transform(img_not_preprocessed)

    img = img.unsqueeze(0)

    img = Variable(img.cuda())
    
    res = fcn(img)
```

```python
%%timeit 
# 100 loops, best of 3: 13.3 ms per loop

benchmark_fcn()
```
